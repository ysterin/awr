```python
import numpy as np 
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
log_files = [Path(f'lander5k/log.txt')] + [Path(f'lander{i}/log.txt') for i in range(2, 6)]
dfs = [pd.read_csv(f, delim_whitespace=True) for f in log_files]
```

```python
plt.figure(figsize=(12, 6))
for i in range(5):
    df = dfs[i]
    plt.plot(df.Samples, df.Test_Return)
plt.xlabel('samples')
plt.ylabel('Test Return')
plt.title('LunarLander-v2')
plt.savefig('figures/lunar_lander.svg')
```
